## Read WIN data

Give the path to the WIN data.  

Channel table is optional.

`encoding` is for channel table file, default is `utf-8`.
In the case of the sample channel table, `encoding` is `euc-jp`.


```python
from pathlib import Path
import wingram

datadir = Path("./WIN_pkg-3.0.11/etc/")
datafp = datadir / "991109.064607"
chtablefp = datadir / "991109.064607.ch"

win = wingram.read(datafp, chtablefp, encoding="euc-jp")
win
```

Actually, the channel table is automatically read when the name of the channel table is (name of WIN data)+.ch and it is in the same directory as the WIN data.

```python
win = wingram.read(datafp, encoding="euc-jp")
win
```

If there is no channel table, the information in channel table set None.
In the following code, the channel table is not read because of encoding error, so only WIN data is read.

```python
win = wingram.read(datafp)
win
```

channel table is hold at win.chtable:

```python
win = wingram.read(datafp, encoding="euc-jp")
win.chtable.iloc[:6]
```

## Raw data to data with unit
Original WIN data only holds integer data.  
To add unit to the data, parameters in channel table is required.


*If you have read channel table*, you can use `win.calibrate` to convert integer data to data with unit.

```python
win = wingram.read(datafp, encoding="euc-jp")
# raw data
win[:3].plot()

# data with unit
win.calibrate()
win[:3].plot()
```

## Basical process

```python
win = wingram.read(datafp, encoding="euc-jp")
win.calibrate()
```

### differentiation and integration

```python
# differential
win.copy().gradient()[0].plot()

# Integration
win.copy().integrate()[0].plot()
```

### filtering

```python
# bandpass filter
win.copy()[0].bandpass(0.1, 1).plot()

# highpass filter
win.copy()[0].bandpass(0.1,None).plot()

# lowpass filter
win.copy()[0].bandpass(None, 10).plot()
```

### demean, detrend

```python
win.select("GNZ","wN").plot()
win.select("GNZ","wN").demean().plot()
win.select("GNZ","wN").detrend().plot()
```

## Obspy to WIN

```python
import obspy
import wingram

st = obspy.read('http://examples.obspy.org/RJOB_061005_072159.ehz.new')
print(st)

print("=================")
win = wingram.from_obspy(st)
print(win)
```

The converted win data is incomplete because obspy.trace does not hold the information of channel number, which leads to the empty `output.ch`.

If you would like to automatically fill in the empty channel numbers, 
you may use `WIN.autofill_ch` method:

```python
print(f"original: {win.ch}") # empty!
win.autofill_ch()
print(f"autofilled: {win.ch}")
```

`obspy.trace` neither holds other parameters for the channel table, so `wingram` set values of `WIN1ch.params` to default.  
However, `obspy.trace.calib` are used to set `ad_bit_step`, which is used to convert raw integer data to data with unit.
